```python
df["Layer2HDD"] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df["Layer2SSD"] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df["Layer2Hybrid"] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df["Layer2Flash_Storage"] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

df['second'] = df['second'].str.replace(r'\D', '')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression

# Define preprocessing steps for categorical and numeric features
categorical_features = ['Company', 'TypeName', 'Cpu brand', 'Gpu brand', 'os']
numeric_features = ['Ram', 'Weight', 'Touchscreen', 'Ips', 'HDD', 'SSD']

# Create a column transformer for both
step1 = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(drop='first'), categorical_features),
    ('num', 'passthrough', numeric_features)
])

# Define the model pipeline
step2 = LinearRegression()

pipe = Pipeline([
    ('preprocessing', step1),
    ('model', step2)
])

# Fit the pipeline
pipe.fit(X_train, y_train)

# Predict and evaluate
y_pred = pipe.predict(X_test)

print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```

```python
step1 = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(drop='first'), categorical_features),
    ('num', 'passthrough', numeric_features)
])

# Define the model pipeline
step2 = KNeighborsRegressor(n_neighbors=3)

pipe = Pipeline([
    ('preprocessing', step1),
    ('model', step2)
])

# Fit the pipeline
pipe.fit(X_train, y_train)

# Predict and evaluate
y_pred = pipe.predict(X_test)

print('R2 score:', r2_score(y_test, y_pred))
print('MAE:', mean_absolute_error(y_test, y_pred))
```
